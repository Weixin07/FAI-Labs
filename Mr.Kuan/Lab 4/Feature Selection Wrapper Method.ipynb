# Feature Selection

## Wrapper Method

```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
df = pd.read_csv('wine.csv')
```

```python
df
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:, 0: 13]
```

```python
X
```

```python
y = df['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
ffs = SequentialFeatureSelector(RandomForestClassifier(n_jobs = -1),
                                k_features = 6,
                                forward = True,
                                floating = False,
                                verbose = 2,
                                scoring = 'accuracy',
                                cv = 5)
```

```python
ffs.fit(X_train, y_train)
```

```python
ffs.k_feature_idx_
```

```python
ffs.k_feature_names_
```

```python
ffs.k_score_
```

```python
pd.DataFrame.from_dict(ffs.get_metric_dict()).T
```

## Topic 2: Backward Elimination

```python
bfe = SequentialFeatureSelector(RandomForestClassifier(n_jobs = -1),
                                k_features = 6,
                                forward = False,
                                floating = False,
                                verbose = 2,
                                scoring = 'accuracy',
                                cv = 5)
```

```python
bfe.fit(X_train, y_train)
```

```python
bfe.k_feature_idx_
```

```python
bfe.k_feature_names_
```

```python
bfe.k_score_
```

```python
bfe = SequentialFeatureSelector(RandomForestClassifier(n_jobs = -1),
                                k_features = (1, 13),
                                forward = False,
                                floating = False,
                                verbose = 2,
                                scoring = 'accuracy',
                                cv = 5)
```

```python
bfe.fit(X_train, y_train)
```

### Topic 3: Exhaustive Feature Selector

```python
df2 = pd.read_csv('sampledata.csv')
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
X = df2.drop('target', axis = 1)
```

```python
y = df2['target']
```

```python
X.head()
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector
```

```python
efs = ExhaustiveFeatureSelector(RandomForestClassifier(n_jobs = -1),
                               min_features = 1,
                               max_features = 4,
                               scoring = 'accuracy',
                               cv = 2,
                               n_jobs = -1)
```

```python
efs.fit(X_train, y_train)
```

```python
efs.best_feature_names_
```

```python
efs.best_score_
```

```python
pd.DataFrame.from_dict(efs.get_metric_dict()).T
```
